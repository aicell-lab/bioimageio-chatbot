```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get the path of the current file
path = os.getcwd()
# get the path of the parent directory
parent_path = os.path.dirname(path)
```

```python
# get current directory
dir_path = os.getcwd()
# load "Knowledge-Retrieval-Evaluation-Results.csv"
score_result = pd.read_csv(dir_path + "/Knowledge-Retrieval-Evaluation_1210_2.csv")
```

```python
# show columns of the score_result
print("score_result.columns: ", score_result.columns)
```

```python
# change the column name "ChatGPT (3.5) Answer" to "ChatGPT3.5-Answer (Without Context)"
score_result.rename(columns={"ChatGPT (3.5) Answer": "ChatGPT3.5-Answer (Without Context)"}, inplace=True)
# save the score_result to csv file
score_result.to_csv(dir_path + "/Knowledge-Retrieval-Evaluation-Results.csv", index=False)
```

```python
# plot 'ChatGPT-3.5 Direct Answer Similarity Score',
    #    'Chatbot Answer Similarity Score', 
    #  'Chatbot Answer Retrieval Precision',
    #    'Chatbot Answer Augmentation Accuracy',
    #    'Chatbot Answer Augmentation Precision',
    #    'Chatbot Answer Augmentation Consistency'
plt.figure(figsize=(5,5))
plt.plot(score_result['GPT-3.5-tubor Answer (Without Context)- Answer Similarity Score'], label='ChatGPT-3.5 Direct Answer Similarity Score')
plt.plot(score_result['GPT4 Direct Answer - Answer Similarity Score'], label='GPT4 Direct Answer Similarity Score')
plt.plot(score_result['BMZ Chatbot Answer - Similarity Score'], label='Chatbot Answer Similarity Score')
plt.xlabel('Question Index')
plt.ylabel('Similarity Score')
plt.legend()
plt.show()
```

```python
# Assuming score_result is a DataFrame with columns containing similarity scores

plt.figure(figsize=(10, 6))

# Plotting histograms with transparency
plt.hist(score_result['GPT-3.5-tubor Answer (Without Context)- Answer Similarity Score'], bins=5, label='ChatGPT-3.5 Direct Answer Similarity Score', alpha=0.5)
plt.hist(score_result['GPT4 Direct Answer - Answer Similarity Score'], bins=5, label='GPT4 Direct Answer Similarity Score', alpha=0.5)
plt.hist(score_result['BMZ Chatbot Answer - Similarity Score'], bins=5, label='BMZ Chatbot Answer Similarity Score', alpha=0.5)

plt.xlabel('Similarity Score')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram of Similarity Scores')
plt.show()
```

```python
# plot histo of three similarity scores, make bar in parallel
plt.figure(figsize=(5,5))
plt.hist(score_result['GPT-3.5-tubor Answer (Without Context)- Answer Similarity Score'], bins=20, label='ChatGPT-3.5 Direct Answer Similarity Score')
plt.hist(score_result['GPT4 Direct Answer - Answer Similarity Score'], bins=20, label='GPT4 Direct Answer Similarity Score')
plt.hist(score_result['BMZ Chatbot Answer - Similarity Score'], bins=20, label='Chatbot Answer Similarity Score')
plt.xlabel('Similarity Score')
plt.ylabel('Frequency')
```

```python
# get the mean of three similarity scores
mean1 = score_result['GPT-3.5-tubor Answer (Without Context)- Answer Similarity Score'].mean()
mean2 = score_result['GPT4 Direct Answer - Answer Similarity Score'].mean()
mean3 = score_result['BMZ Chatbot Answer - Similarity Score'].mean()
labels = ['ChatGPT-3.5', 'GPT4', 'BMZ Chatbot']
# plot 3 mean values in the plot
plt.bar(range(3), [mean1, mean2, mean3], color='orange', label='Mean')
# Adding labels and title
# plt.xlabel('Values')
# plt.ylabel('Magnitude')
plt.title('Similarity Scores (0-5)')
plt.ylim(3,4.4)
# Adding text annotations for each value
for i, value in enumerate([mean1, mean2, mean3]):
    plt.text(i, value+0.1, str(round(value,2)))
# Setting custom labels for x-axis
plt.xticks(range(3), labels)
# add grid
# plt.grid()
plt.show()
```
